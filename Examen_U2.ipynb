#**Hipótesis**
Personas con una mayor edad, masa corporal y circunferencia de cintura tienden a tener niveles más bajos de actividad física, lo cual en conjunto incrementa significativamente el riesgo de hipertensión.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import statsmodels.api as sm
import matplotlib.pyplot as plt

X = df[['edad','valor_trigliceridos', 'peso', 'estatura', 'medida_cintura', 'segundamedicion_peso', 'circunferencia_de_la_pantorrilla', 'segundamedicion_cintura', 'masa_corporal']]
Y = df['riesgo_hipertension']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=42)

X_train_constante = sm.add_constant(X_train)
X_test_constante = sm.add_constant(X_test)

modelo = sm.OLS(Y_train, X_train_constante).fit()


Yc = modelo.predict(X_test_constante)

plt.scatter(Y_test, Yc)

r2 = r2_score(Y_test, Yc)
print(f'Coeficiente de determinacion: {r2: 0.4f}')
```

El resultado del coeficiente de determinación significa que Y está sujeto a X con un 38% de variación.

```python
b0, b1, b2, b3, b4, b5, b6, b7, b8, b9 = modelo.params

Fun = lambda X1, X2, X3, X4, X5, X6, X7, X8, X9: b0 + b1 * X1 + b2 * X2 + b3 * X3 + b4 * X4 + b5 * X5 + b6 * X6 + b7 * X7 + b8 * X8 + b9 * X9

Fun(50, 150, 70, 150, 90, 60, 33, 100, 20)
```

```python
from statsmodels.stats.api import het_breuschpagan
_, valor_p_bp, _, _ = het_breuschpagan(modelo.resid, X_train_constante)
print(f'Valor-p de Breusch-Pagan: {valor_p_bp: 0.4f}')
```

```python
from statsmodels.formula.api import ols

moedo_2 = ols(formula = 'riesgo_hipertension ~ edad + valor_trigliceridos + peso + estatura + medida_cintura + segundamedicion_peso + circunferencia_de_la_pantorrilla + segundamedicion_cintura + masa_corporal', data = df).fit()
tabla_anova = sm.stats.anova_lm(moedo_2, typ = 2)  # ANOVA de una via
tabla_anova
```

##**Conclusión**
Los resultados obtenidos con el apoyo del modelo de regresión lineal múltiple, nos ayudan a comprobar nuestra hipótesis inicial, ya que con ello podemos entender cómo cada uno de estos factores (edad, valor trigliceridos, peso, estatura, medida cintura, segunda medicion peso, circunferencia de la pantorrilla, segunda medicion cintura y masa corporal) influyen en los individuos de tal manera que pueden provocar el riesgo de hipertensión.

Este análisis demuestran que estos diversos factores tienen un impacto significativo en los individuos. Y es muy importante tomar en cuenta los datos obtenidos, ya que con este pronóstico nos será más facil poder encontrar la manera de evitar que suceda este problema en las personas, con ayuda de una mejor alimentación y haciendo ejercicio.
