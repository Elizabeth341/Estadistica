##**Introducci√≥n**

En el √°mbito educativo, el rendimiento acad√©mico es un tema que ha llevado a realizar muchas investigaciones debido al impacto que tiene en los estudiantes, ya que existen muchos factores que afectan las calificaciones que obtienen.

Algunos de estos factores son el n√∫mero de cursos y el tiempo de estudio, al analizar estos elementos, podemos predecir c√≥mo se desempe√±ar√°n los alumnos y aplicar estrategias que les ayuden a aprender y a mejorar sus calificaciones.


El prop√≥sito de este proyecto es analizar un conjunto de datos con el fin de identificar los factores y la manera que influyen en las calificaciones obtenidas por los estudiantes.


##**Hip√≥tesis**

Existe una relaci√≥n positiva entre el tiempo de estudio y las calificaciones de los estudiantes, debido a que entre m√°s tiempo inviertan estudiando, mejores ser√°n sus resultados.


##**Marco te√≥rico**

El entorno educativo en la actualidad ha tomado un papel fundamental en la vida de las personas, debido al aprendizaje que se puede adquirir para alcanzar metas futuras. Cada vez m√°s, hay personas que aprovechan la oportunidad de seguir estudiando para alcanzar lo que desean.

A√∫n as√≠, es importante mencionar que existen elementos que intervienen en el aprendizaje, lo que lleva a que las personas no obtengan la productividad que esperan.

Diversas investigaciones han demostrado que el rendimiento acad√©mico de los estudiantes puede verse influido por varios factores. Algunos de estos son:

*   **N√∫mero de cursos:** El n√∫mero de cursos que un estudiante toma influye mucho, ya que entre m√°s cursos tenga ser√° m√°s pesada su carga acad√©mica, lo que afecta su rendimiento.
*   **Tiempo de estudio:** El tiempo que un alumno invierte estudiando influye demasiado en sus calificaciones finales, ya que una persona que estudia mucho es probable que tenga mejores resultados.


Es por eso que, el principal objetivo de este an√°lisis es descubrir como intervienen estos factores en las calificaciones de los estudiantes, es decir si uno influye al otro. Para esto la regresi√≥n lineal simple es √∫til ya que nos ayudar√° a saber la relaci√≥n que existe entre estas variables.

El modelo de regresi√≥n lineal simple es una f√≥rmula matem√°tica que trata de encontrar la relaci√≥n que existe entre dos variables usando una linea recta.
Est√° representada como: Y = Œ≤0 + Œ≤1X + œµ


Los **supuestos** del modelo son los siguientes:


*   **Linealidad**:Existe relaci√≥n lineal entre las dos variables
*   **Independencia**: Los datos no dependen unos de otros.
*   **Homoscedasticidad**: Los puntos se distribuyen con la misma dispersi√≥n, tienen la misma varianza.


El **coeficiente de correlaci√≥n** nos servir√° para medir la relaci√≥n entre las variables

Se representa como: -1 ‚â§ r ‚â§ 1

Donde:

*   1 = Relaci√≥n fuerte y positiva.
*  -1 = Relaci√≥n fuerte y negativa.
*   0 = Sin relaci√≥n lineal.

El **coeficiente de determinaci√≥n** nos dir√° que porcentaje se puede explicar con la variable que usamos para hacer la predicci√≥n

Se representa como: 0 ‚â§ r2 ‚â§ 1

Donde:

*   1 = Predicci√≥n perfecta.
*   0 = No hay relaci√≥n.

El **test de Shapiro** nos ayudar√° a comprobar si los datos siguen una distribuci√≥n normal.

Las hip√≥tesis son las siguientes:

*   ùêª0: Los datos siguen una distribuci√≥n normal.
*   ùêª1: Los datos no siguen una distribuci√≥n normal.

El **test de Breusch-Pagan** detectar√° si hay heteroscedasticidad en el modelo.

Las hip√≥tesis son las siguientes:

*   ùêª0: Hay homoscedasticidad.
*   ùêª1: Hay heteroscedasticidad.

El **ANOVA** nos servir√° para verificar si la pendiente de la regresi√≥n es estad√≠sticamente significativa.

Las hip√≥tesis son las siguientes:

*   ùêª0: Œ≤1 = 0.
*   ùêª1: Œ≤1 ‚â† 0.

Realizar este an√°lisis nos dar√° informaci√≥n para saber si el tiempo que los alumnos dedican al estudio tiene un efecto significativo en sus calificaciones.


##**Metodolog√≠a**

1. Seleccionar las Variables

*   Variable independiente (X): Tiempo de estudio (Time study)
*   Variable dependiente (Y): Calificaciones (Marks)

2. Crear un gr√°fico de dispersi√≥n
Se realiza un gr√°fico de dispersi√≥n para visualizar la relaci√≥n entre el tiempo de estudio y las calificaciones.

3. Calcular el Coeficiente de Correlaci√≥n de Pearson
Se calcula el coeficiente de correlaci√≥n (r), que mide la fuerza de la relaci√≥n lineal entre las dos variables. Un valor cercano a 1 indica una fuerte relaci√≥n positiva, mientras que un valor cercano a -1 indica una fuerte relaci√≥n negativa, en cambio si el valor es 0 indica que no hay relaci√≥n.

4. Calcular el Coeficiente de Determinaci√≥n
El coeficiente de determinaci√≥n (r¬≤) se calcula para conocer qu√© porcentaje de la variabilidad de las calificaciones (Y) se puede explicar por el tiempo de estudio (X).

5. Calcular la Recta de Regresi√≥n Lineal
Se calcula la recta de regresi√≥n lineal, que representa la l√≠nea recta que mejor se adapta a los puntos y con ella se observa la relaci√≥n entre el tiempo de estudio y las calificaciones.

6. Calcular el Intervalo de Confianza del 95% para b1
Se calcula el intervalo de confianza del 95% para el valor de b1, esto proporciona un rango dentro del cual se espera que se encuentre el valor verdadero de la pendiente.

7. Calcular los residuales
Se calculan los residuales, estos nos permiten verificar los supuestos como la linealidad, homoscedasticidad y la normalidad de errores.

8. Calcular la prueba de Shapiro
Se realiza la prueba de Shapiro para verificar si los residuales siguen una distribuci√≥n normal, lo cual es un supuesto importante en la regresi√≥n lineal.

9. Calcular la prueba de Breusch-Pagan
Se realiza la prueba de Breusch-Pagan para identificar si existe heteroscedasticidad, es decir, si existe una varianza diferente.

10. Realizar una tabla ANOVA
Se realiza una tabla ANOVA para verificar si el modelo de regresi√≥n es estad√≠sticamente significativo, lo que eval√∫a si el tiempo de estudio tiene un impacto real en las calificaciones.
